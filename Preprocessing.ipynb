## Valeurs manquantes

```python
# Méthode pour calculer le nombre de valeurs manquantes pour chaque variable explicative
def valeursManquantes(df):
        valeurs = df.isnull().sum()
        valeursPourcent = 100 * df.isnull().sum() / len(df)
        table = pd.concat([valeurs, valeursPourcent], axis=1)
        
        # On rennomme les colonnes
        tableRenommer = table.rename(columns = {0 : "Valeurs Manquantes", 1 : "% des valeurs totales"})
        tableRenommer = tableRenommer[tableRenommer.iloc[:,1] != 0].sort_values("% des valeurs totales", ascending=False).round(1)
        
        return tableRenommer

valeursManquantes(data)
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

new_column_platform = encoder.fit_transform(data["Platform"].values)
new_column_genre = encoder.fit_transform(data["Genre"].values)
new_column_publisher = encoder.fit_transform(data["Publisher"].values)
new_column_developer = encoder.fit_transform(data["Developer"].values)
new_column_rating = encoder.fit_transform(data["Rating"].values)

encoded_data = data.copy()

encoded_data["Platform"] = new_column_platform
encoded_data["Genre"] = new_column_genre
encoded_data["Publisher"] = new_column_publisher
encoded_data["Developer"] = new_column_developer
encoded_data["Rating"] = new_column_rating

encoded_data.head()
```

```python
y = encoded_data['Global_Sales']
encoded_data = encoded_data.drop(columns=['Global_Sales', 'Name'])

encoded_data_1 = encoded_data.drop(columns=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'])
X1 = encoded_data_1.to_numpy()
encoded_data_2 = encoded_data.drop(columns=['Platform', 'Year_of_Release', 'Genre', 'Publisher', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Developer', 'Rating'])
X2 = encoded_data_2.to_numpy()
```

## MinMaxScaler

```python
from  sklearn.preprocessing import MinMaxScaler
```

```python
X1 = MinMaxScaler().fit_transform(X1)
X2 = MinMaxScaler().fit_transform(X2)
```

## Linear Reg

```python
from sklearn.linear_model import LinearRegression
```

```python
reg1 = LinearRegression().fit(X1, y)
reg2 = LinearRegression().fit(X2, y)
```

```python
weights1 = reg1.coef_
weights1 = (weights1 - np.min(weights1))/np.ptp(weights1)

weights2 = reg2.coef_
weights2 = (weights2 - np.min(weights2))/np.ptp(weights2)
```

```python
columns1 = encoded_data_1.columns

columns2 = encoded_data_2.columns
```

```python
d1 = {'columns': columns1, 'weights': weights1}
d2 = {'columns': columns2, 'weights': weights2}
df1 = pd.DataFrame(data=d1)
df2 = pd.DataFrame(data=d2)

print(df1)
print(df2)
```

```python
df1.to_csv('./data/weights_nonsale.csv', index=False)
df2.to_csv('./data/weights_sale.csv', index=False)
```
